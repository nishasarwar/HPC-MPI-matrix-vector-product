```python
import matplotlib.pyplot as plt
```

```python
# Read the data
file_path = "strong_scaling_results.txt"
ranks = []
avg_compute = []
avg_comm = []

with open(file_path, 'r') as f:
    lines = f.readlines()[1:]  # skip header
    for line in lines:
        parts = line.strip().split()
        if len(parts) >= 3:
            ranks.append(int(parts[0]))
            avg_compute.append(float(parts[1]))
            avg_comm.append(float(parts[2]))
```

```python
# Calculate speedup
baseline = avg_compute[0]  # compute time at 1 rank
speedup = [baseline / t for t in avg_compute]

# Plot 1: Speedup
plt.figure(figsize=(8, 5))
plt.plot(ranks, speedup, marker='o', linestyle='-', color='green')
plt.xscale('log', base=2)
plt.xticks(ranks)
plt.xlabel("Number of Ranks (log scale)")
plt.ylabel("Speedup")
plt.title("Strong Scaling Speedup vs. Number of Ranks")
plt.grid(True, which="both", ls="--")
plt.tight_layout()
plt.savefig("q2-speedup-plot.png")
plt.show()
```

```python
# Plot 2: Communication Time
plt.figure(figsize=(8, 5))
plt.plot(ranks, avg_comm, marker='s', linestyle='-', color='red')
plt.xscale('log', base=2)
plt.xticks(ranks)
plt.xlabel("Number of Ranks (log scale)")
plt.ylabel("Average Communication Time (s)")
plt.title("Communication Cost vs. Number of Ranks")
plt.grid(True, which="both", ls="--")
plt.tight_layout()
plt.savefig("q3-communication-plot.png")
plt.show()
```
